```python
from faster_whisper import WhisperModel
import soundfile as sf
import os

# 指定模型檔案的路徑
model_local_path = "/content/models--andybi7676--cool-whisper/snapshots/9243ae0ca0cff575d2ca8a5c5698e232bf47821c"

# 指定音檔路徑
audio_fpath = "/content/2.wav"  # 修改為您的音檔路徑
if not os.path.exists(audio_fpath):
    raise FileNotFoundError(f"Audio file '{audio_fpath}' does not exist. Please upload the file to this path.")

# 自動生成 SRT 檔名
base_name = os.path.splitext(os.path.basename(audio_fpath))[0]
srt_file_path = f"/content/{base_name}.srt"

# 載入模型
model = WhisperModel(model_local_path, device="cuda", compute_type="float16")

# 進行音檔轉錄
segments, info = model.transcribe(
    audio_fpath,
    beam_size=5,
    language="zh",  # 指定語言為中文
    condition_on_previous_text=True
)

# 將結果儲存為 .srt
def format_time(seconds):
    """將秒數轉換為 SRT 格式的時間"""
    millis = int((seconds - int(seconds)) * 1000)
    time = f"{int(seconds // 3600):02}:{int(seconds % 3600 // 60):02}:{int(seconds % 60):02},{millis:03}"
    return time

with open(srt_file_path, "w", encoding="utf-8") as srt_file:
    for idx, segment in enumerate(segments, start=1):
        start_time = format_time(segment.start)
        end_time = format_time(segment.end)
        srt_file.write(f"{idx}\n")
        srt_file.write(f"{start_time} --> {end_time}\n")
        srt_file.write(f"{segment.text}\n\n")

print(f"SRT 檔案已儲存至 {srt_file_path}")

# 下載 SRT 文件
from google.colab import files
files.download(srt_file_path)
```

```python
from faster_whisper import WhisperModel
import os
import csv
import requests
import re

# 設定模型參數
model_local_path = "/content/models--andybi7676--cool-whisper/snapshots/9243ae0ca0cff575d2ca8a5c5698e232bf47821c"
audio_folder = "/content/unzipped_files/kaldi-taiwanese-asr/test"  # 音檔資料夾
output_csv = "/content/test.csv"  # 輸出 CSV 檔案路徑

# 台羅拼音轉換函式
def convert_to_taigi(text):
    url = "http://tts001.iptcloud.net:8804/display"
    params = {"text0": text}
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.text.strip()
    else:
        return f"[Error] {text}"

# 提取檔名中的數字，用於排序
def extract_number(filename):
    match = re.search(r"(\d+)\.wav$", filename)
    return int(match.group(1)) if match else float('inf')

# 初始化 Whisper 模型
model = WhisperModel(model_local_path, device="cuda", compute_type="float16")

# 取得所有音檔，並根據檔名中的數字排序
audio_files = sorted(
    [f for f in os.listdir(audio_folder) if f.endswith(".wav")],
    key=extract_number
)

# 轉錄並寫入 CSV
with open(output_csv, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["id", "text"])  # 寫入 CSV 標題

    for audio_file in audio_files:
        audio_id = extract_number(audio_file)  # 提取檔案 ID
        audio_path = os.path.join(audio_folder, audio_file)
        print(f"正在處理 {audio_path} ...")

        # 轉錄音檔
        segments, _ = model.transcribe(
            audio_path,
            beam_size=5,
            language="zh",
            condition_on_previous_text=True
        )

        # 拼接轉錄文字
        chinese_text = "".join(segment.text for segment in segments)

        # 轉換成台羅拼音
        tai_lo_text = convert_to_taigi(chinese_text)

        # 寫入 CSV 檔案
        writer.writerow([audio_id, tai_lo_text])
        print(f"已完成第 {audio_id} 個檔案: {tai_lo_text}")

print(f"所有音檔處理完畢，結果已儲存至 {output_csv}")
```

```python
import pandas as pd
import re

# 定義移除音調數字的正規化函式
def normalize_taigi_pinyin(text):
    """
    移除台羅拼音中的音調數字，例如 'li2 be7 e5' -> 'li be e'
    """
    return re.sub(r'\d+', '', text)

# 讀取 CSV 檔案
input_csv = "/content/test.csv"  # 原始結果檔案
output_csv = "/content/test_normalized.csv"  # 輸出正規化檔案

# 讀取 CSV 並進行正規化
df = pd.read_csv(input_csv)

# 確保檔案有 'text' 欄位
if 'text' in df.columns:
    df['text'] = df['text'].apply(normalize_taigi_pinyin)
else:
    raise ValueError("CSV 檔案中找不到 'text' 欄位！")

# 儲存正規化後的結果
df.to_csv(output_csv, index=False, encoding='utf-8')
print(f"拼音正規化完成，結果已儲存至：{output_csv}")

# 預覽輸出檔案
print("正規化後的前5筆資料：")
print(df.head())
```

```python
import pandas as pd
import re

# 定義處理函式：替換 '-' 和 ', ' 並正規化多個空白
def clean_text(text):
    """
    1. 將 '-' 替換為 ' '
    2. 將 ', ' 替換為 ' '
    3. 將多個空白 '  ' 替換為單一空白 ' '
    """
    text = text.replace("-", " ")  # 將 '-' 換成空白
    text = text.replace(", ", " ")  # 將 ', ' 換成空白
    text = re.sub(r'\s+', ' ', text)  # 將多個空白替換成單一空白
    return text.strip()  # 移除首尾多餘的空白

# 讀取 CSV 檔案
input_csv = "/content/test_cleaned.csv"  # 輸入檔案路徑
output_csv = "/content/test_cleaned_2.csv"  # 輸出檔案路徑

# 讀取 CSV 並清理 text 欄位
df = pd.read_csv(input_csv)

# 檢查 text 欄位是否存在
if 'text' in df.columns:
    df['text'] = df['text'].apply(clean_text)  # 套用處理函式
else:
    raise ValueError("CSV 檔案中找不到 'text' 欄位！")

# 儲存處理後的結果
df.to_csv(output_csv, index=False, encoding='utf-8')
print(f"文本正規化完成，結果已儲存至：{output_csv}")

# 預覽輸出檔案
print("處理後的前5筆資料：")
print(df.head())
```

```python
import pandas as pd
import re

# 定義處理函式：清理文本內容
def clean_text(text):
    """
    1. 將 '-' 替換為 ' '
    2. 將 ', ' 替換為 ' '
    3. 將多個空白 '  ' 替換為單一空白 ' '
    """
    if pd.isna(text):  # 處理 NaN 值
        return ""
    text = text.replace("-", " ")  # 將 '-' 換成空白
    text = text.replace(", ", " ")  # 將 ', ' 換成空白
    text = re.sub(r'\s+', ' ', text)  # 將多個空白替換成單一空白
    return text.strip()  # 移除首尾多餘空白

# 讀取 CSV 檔案
input_csv = "/content/test_cleaned_2.csv"  # 輸入檔案路徑
output_csv = "/content/test_fixed.csv"  # 輸出檔案路徑

# 讀取 CSV，處理錯位欄位
df = pd.read_csv(input_csv, header=0)

# 檢查是否有第三欄，並將其覆蓋到第二欄
if len(df.columns) > 2:
    print("發現第三欄，進行修復處理...")
    df.iloc[:, 1] = df.apply(lambda row: row.iloc[2] if pd.notna(row.iloc[2]) else row.iloc[1], axis=1)

# 清理第二欄文本內容
df.iloc[:, 1] = df.iloc[:, 1].apply(clean_text)

# 保留第一欄和第二欄，並重新命名欄位
df = df.iloc[:, [0, 1]]
df.columns = ['id', 'text']

# 儲存修復後的檔案
df.to_csv(output_csv, index=False, encoding='utf-8')
print(f"修復完成，結果已儲存至：{output_csv}")

# 預覽輸出檔案
print("修復後的前5筆資料：")
print(df.head())
```

```python
import pandas as pd

# 檔案路徑
input_csv = "/content/test_fixed.csv"  # 輸入檔案路徑
output_csv = "/content/test_fixed_2.csv"  # 輸出檔案路徑

# 讀取 CSV 檔案
df = pd.read_csv(input_csv)

# 定義函式：移除 ? 字元
def remove_question_mark(text):
    if pd.isna(text):  # 處理 NaN
        return ""
    return text.replace("?", "")

# 對 text 欄位進行處理
if 'text' in df.columns:
    df['text'] = df['text'].apply(remove_question_mark)
else:
    raise ValueError("CSV 檔案中找不到 'text' 欄位！")

# 儲存結果
df.to_csv(output_csv, index=False, encoding='utf-8')
print(f"所有 '?' 字元已移除，結果儲存至：{output_csv}")

# 預覽處理後的檔案
print("處理後的前5筆資料：")
print(df.head())
```
